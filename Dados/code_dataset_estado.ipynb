# Dataset Estado COVID-19

Paraíba

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv")
df['date'] = pd.to_datetime(df['date'])
print(df.columns)

data_PB = df.query("state == 'PB'")[['date', 'totalCases', 'deaths', 'suspects', 'recovered', 'tests', 'vaccinated', 'vaccinated_second']]
data_PB['activeCases'] = data_PB['totalCases'] - data_PB['deaths'] - data_PB['recovered']
data_PB['newVaccinated'] = data_PB['vaccinated'].diff()
data_PB['newVaccinated_second'] = data_PB['vaccinated_second'].diff()

data_PB
```

```python
data_PB.plot(x="date", y="deaths",xlabel="Data",ylabel="Número de Mortos",label="Número Total de Mortos da Paraíba")
```

```python
data_PB.plot(x="date", y="totalCases",xlabel="Data",ylabel="Número de Casos",label="Número Total de Casos da Paraíba")
```
